```python
#BAR PLOT
#A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

import matplotlib.pyplot as plt
x=["python","c","c++","java","R"]
y=[89,70,80,85,45]
plt.xlabel("language",fontsize=15)
plt.ylabel("rate",fontsize=15)
plt.title("anisha",fontsize=15)
plt.bar(x,y,color="y",width=0.8,align="edge",edgecolor="b",linewidth=2,linestyle=":",alpha=0.5,label="popularity")
plt.legend()
plt.show()

# bar() function to draw bar graphs
# color to set the color of the bars
# width to set the width of the bars
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java","R"]
y=[89,70,80,85,45]
z=[20,30,40,50,60]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("rate",fontsize=15)
plt.title("anisha",fontsize=15)

plt.bar(p,y,width,color="y",label="popularity")
plt.bar(p1,z,width,color="r",label="popularity1")

        #position,name,roation
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
#If you want the bars to be displayed horizontally instead of vertically, use the barh() function

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```
