```python
#FILL_BETWEEN_PLOT
#fill area between two horizontal curves. 
#Two points (x, y1) and (x, y2) define the curves. this creates one or more polygons describing the filled areas. 
#The ‘where’ parameter can be used to selectively fill some areas

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
# plt.fill_between(x=[2,4],y1=2,y2=4)
plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.4)
plt.title("fill_between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")         
plt.show()
```
