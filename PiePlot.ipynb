```python
#PIE PLOT
#A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data.

import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.3,0.0,0.0,0.0]
c=["g","b","m","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1,labeldistance=1.1,startangle=0
       ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"m"},center=(3,3)
       ,rotatelabels=False)

plt.title("Pie plot",fontsize=20)
plt.legend(loc=2)
plt.show()
```

```python
#DOT PIECHART
plt.pie([1])
plt.show()
```

```python
#DONUT PIECHART

import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]
c=["g","r","b","m"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1)

plt.show()
```
