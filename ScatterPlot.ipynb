```python
#SCATTER PLOT
#A scatter plot is a diagram where each value in the data set is represented by a dot. The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for the values of the y-axis
#matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)[source]

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,3,6,8,11,12,14,18]
# colors=["g","r","m","y","b","r","m","b"]
colors=[100,20,30,45,50,60,70,80]
sizes=[100,150,200,250,300,350,400,450,]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.4,marker="*",edgecolor="g",linewidth=2,cmap="viridis")
t=plt.colorbar()
t.set_label("colorbar",fontsize=20)
plt.title("scatter plot",fontsize=20,)
plt.xlabel("days",fontsize=20)
plt.ylabel("no.",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,3,6,8,11,12,14,18]
z=[3,4,5,6,7,8,9,2]
# colors=["g","r","m","y","b","r","m","b"]
colors=[100,20,30,45,50,60,70,80]
sizes=[100,150,200,250,300,350,400,450,]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.4,marker="*",edgecolor="g",linewidth=2,cmap="viridis")
plt.scatter(x,z,c="r",s=sizes)
t=plt.colorbar()
t.set_label("colorbar",fontsize=20)
plt.title("scatter plot",fontsize=20,)
plt.xlabel("days",fontsize=20)
plt.ylabel("no.",fontsize=20)
plt.show()
```
