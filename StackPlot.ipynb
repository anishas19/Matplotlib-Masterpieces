```python
#STACK PLOT(also known as area plot)
#Stackplots are generated by plotting different datasets vertically on top of one another rather than overlapping with one another. 

import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,3]
area2=[3,3,1,5,4]
area3=[2,1,4,6,3]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["g","m","r"],#baseline="zero" or "wiggle" or "sym"
             )
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.grid()
plt.legend(loc=2)
plt.show()
```
