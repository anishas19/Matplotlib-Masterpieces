```python
#STEM PLOT
#A Stem plot plots vertical lines at each x position covered under the graph from the baseline to y, and places a marker there.

import matplotlib.pyplot as plt
x=[1,2,3,4,5,3]
y=[2,2,5,7,3,3]

plt.stem(x,y,linefmt="--",markerfmt="ro",bottom=1,basefmt="g",label="python"
         ,orientation="horizontal")
plt.legend()
plt.show()
```
