```python
#STEP PLOT
#This method uses a standard plot with a step drawstyle: The x values are the reference positions and steps extend left/right/both directions depending on where.

import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="python")
plt.title("step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend(loc=2)
plt.show
```
