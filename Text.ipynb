```python
#WORKING WITH TEXT
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,1,6,5,4]

plt.plot(x,y)
plt.title("anisha",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("python",fontsize=15)
plt.text(2,5,"py",fontsize=15,style="italic",bbox={"facecolor":"green"})
plt.annotate("python",xy=(2,1),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))
plt.legend(["up"],loc=6,facecolor="y",edgecolor="g",framealpha=0.4,shadow=True)
plt.show()
```
